```python
import csv

def parse_course_list(list_name, year):
    with open(f'./engAdvisorHtml/{list_name}__{year}.html', 'r') as f:
        html = f.read()
    soup = BeautifulSoup(html, 'html.parser')
    # Select the course faculty code
    faculty_codes = [td.text.strip() for td in soup.select('table.content td:nth-of-type(1)')]
    print(faculty_codes)
    # Select the course numbers
    course_numbers = [td.text.strip() for td in soup.select('table.content td:nth-of-type(2)')]
    print(course_numbers)
    
    with open(f'./engAdvisorCsv/{list_name}__{year}.csv', 'w') as f:
        writer = csv.writer(f)
        writer.writerow(['faculty_code', 'course_number'])
        for faculty_code, course_number in zip(faculty_codes[1:], course_numbers):
            writer.writerow([faculty_code, course_number])

for year in years:
    for list_name in lists:
        parse_course_list(list_name, year)
```
